```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
```

```python
train_df = pd.read_csv('X_train_scaled2.csv')
train_df
```

```python
genre_hierarchy = {
    'Rock and Alternative': ['album rock', 'alternative rock', 'art rock', 'blues rock', 'classic rock', 'folk rock', 'hard rock', 'indie rock', 'modern alternative rock', 'modern rock', 'psychedelic rock', 'rock', 'rock en espanol', 'rock-and-roll', 'roots rock'],
    'Pop and Dance': ['brill building pop', 'bubblegum pop', 'dance pop', 'dance rock', 'disco', 'edm', 'electropop', 'filmi', 'hip pop', 'indie poptimism', 'new romantic', 'new wave', 'new wave pop', 'pop', 'pop dance', 'pop punk', 'pop rap', 'pop rock', 'post-teen pop', 'show tunes', 'soft rock', 'stomp and holler', 'traditional folk', 'tropical house'],
    'Hip Hop and Rap': ['alternative hip hop', 'dirty south rap', 'gangster rap', 'hardcore hip hop', 'hip hop', 'latin hip hop', 'rap', 'southern hip hop', 'trap', 'trap latino', 'underground hip hop'],
    'Jazz and Blues': ['bebop', 'big band', 'cool jazz', 'hard bop', 'jazz', 'jazz funk', 'lounge', 'mellow gold', 'neo soul', 'quiet storm', 'soul jazz', 'vocal jazz'],
    'Country and Americana': ['ccm', 'contemporary country', 'country', 'country road', 'country rock', 'folk', 'indie folk', 'new americana', 'regional mexican'],
    'Soul, R&B, and Related Genres': ['classic soul', 'funk', 'motown', 'neo mellow', 'r&b', 'soul', 'urban contemporary'],
    'Latin, Tropical, and Regional Varieties': ['corrido', 'latin', 'latin pop', 'nuevo regional mexicano', 'reggaeton', 'tropical']
}
```

```python
main_train_df = pd.read_csv('/Users/saranyak/Desktop/GT/Fall2023/ComputationalDataAnalytics/Project/GenreClassification/Preprocessing/Output/NotEncoded_Genres_ArtistsEncoded/p%ofgenre/Train.csv')
```

```python
main_train_df
```

```python
train_df1 = pd.merge(main_train_df, train_df[['genre']], left_index=True, right_index=True)
```

```python
train_df1
```

```python
X_train = train_df1.drop(['genre', 'genres'], axis=1)
y_train = train_df1['genres']

# Initialize and train a decision tree classifier
classifier = DecisionTreeClassifier()
classifier.fit(X_train, y_train)
```

```python
val_df=pd.read_csv('X_valid_scaled2.csv')
valid_df = pd.merge(main_train_df, val_df[['genre']], left_index=True, right_index=True)
valid_df
```

```python
X_val = valid_df.drop(['genre', 'genres'], axis=1)
y_val = valid_df['genres']
```

```python
y_val_pred = classifier.predict(X_val)

# Calculate accuracy on the validation set
val_accuracy = accuracy_score(y_val, y_val_pred)
print(f'Validation Accuracy: {val_accuracy:.2f}')

# Display classification report for validation set
print('\nValidation Classification Report:')
print(classification_report(y_val, y_val_pred))
```

```python
test_df1=pd.read_csv('X_test.csv')
test_df = pd.read_csv('/Users/saranyak/Desktop/GT/Fall2023/ComputationalDataAnalytics/Project/GenreClassification/Preprocessing/Output/NotEncoded_Genres_ArtistsEncoded/p%ofgenre/Test.csv')
test_df = pd.merge(test_df, test_df1[['genre']], left_index=True, right_index=True)
test_df
```

```python
X_test = test_df.drop(['genre', 'genres'], axis=1)
y_test = test_df['genres']
```

```python
y_test_pred = classifier.predict(X_test)

# Calculate accuracy on the test set
test_accuracy = accuracy_score(y_test, y_test_pred)
print(f'\nTest Accuracy: {test_accuracy:.2f}')

# Display classification report for test set
print('\nTest Classification Report:')
print(classification_report(y_test, y_test_pred))
```

```python
from xgboost import XGBClassifier
```

```python
level1_classifier = XGBClassifier()
level1_classifier.fit(X_train, y_train)

# Predictions for the first level on validation set
y_val_pred_level1 = level1_classifier.predict(X_val)
y_test_pred_level1 = level1_classifier.predict(X_test)
```

```python
# Loop through each genre category and train a new XGBoost classifier for the subcategories
for genre_category in genre_hierarchy.keys():
    # Filter the data for the specific genre category
    subcategory_train_df = train_df1[train_df1['genre'] == genre_category]
    
    # Features and labels for the subcategory
    X_subcategory = subcategory_train_df.drop(['genres', 'genre'], axis=1)
    y_subcategory = subcategory_train_df['genres']
    
    # Initialize and train XGBoost classifier for the subcategory
    subcategory_classifier = XGBClassifier()
    subcategory_classifier.fit(X_subcategory, y_subcategory)

    # Use the predictions of the first-level model as features for the second-level model on validation set
    X_val_subcategory = X_val.copy()
    X_val_subcategory['prediction_level1'] = y_val_pred_level1
    
    # Ensure only relevant features are included
    X_val_subcategory = X_val_subcategory[X_subcategory.columns]

    # Predictions for the subcategory
    y_val_pred_subcategory = subcategory_classifier.predict(X_val_subcategory)

    # Display results for the subcategory on validation set
    print(f'\nClassification Report for {genre_category} on Validation Set:')
    print(classification_report(y_val, y_val_pred_subcategory))
    
    X_test_subcategory = X_test.copy()
    X_test_subcategory['prediction_level1'] = y_test_pred_level1

    
    # Ensure only relevant features are included
    X_test_subcategory = X_test_subcategory[X_subcategory.columns]

    # Predictions for the subcategory
    y_test_pred_subcategory = subcategory_classifier.predict(X_test_subcategory)

    # Display results for the subcategory on test set
    print(f'\nClassification Report for {genre_category} on Test Set:')
    print(classification_report(y_test, y_test_pred_subcategory))
```
